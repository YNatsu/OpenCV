```python
import cv2
import matplotlib.pyplot as plt
import numpy as py
%matplotlib inline

# 读取彩色图像
img = cv2.imread(filename='resource/exp.jpg',flags=1)
```

```python
# BGR 转 RGB
img = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2RGB,)
plt.imshow(img)
```

```python
# 读取黑白图像
img = cv2.imread('resource/exp.jpg', 0)
img.shape
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img,cmap='gray')
```

```python
# 保存图像
cv2.imwrite(filename='resource/exp_gray.jpg', img=img)
```

```python
# 通道提取
img = cv2.imread(filename='resource/exp.jpg',flags=1)
b, g, r = cv2.split(img)
```

```python
b.shape
```

```python
g.shape
```

```python
r.shape
```

```python
# 通道融合
img = cv2.merge([b, r, g])
img.shape
```

```python
# 图像融合
img = cv2.imread(filename='resource/exp.jpg',flags=1)
img1 = cv2.imread(filename='resource/exp1.jpg',flags=1)
```

```python
img_resize = cv2.resize(img, (2400, 1352))
```

```python
img_resize.shape
```

```python
img1.shape
```

```python
res = cv2.addWeighted(img_resize, 0.4, img1, 0.6 , 0)
plt.imshow(res)
```

```python
# 图像拉伸
img_f = cv2.resize(img, (0, 0), fx=2, fy=1)
plt.imshow(img_f)
```
